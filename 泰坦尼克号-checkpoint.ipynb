```python
#数据读取
import pandas as pd
import numpy as np
# 使用read_csv()方法进行读取
#加入index是为了一会数据集的整合与拆分
train = pd.read_csv(r'C:/Users/LENOVO/Downloads/train (1).csv',index_col=0)
test  = pd.read_csv(r'C:/Users/LENOVO/Downloads/test (1).csv',index_col=0)
train.head()
```

```python
#因为存在很多列存在缺省值，测试集和训练集都有
#如果测试集没有处理，则无法使用模型进行预测
#所以需要将测试集和训练集合并处理
#但是测试集中没有生存值，所以需要首先处理训练集的生存值
#通过条形图观察生存值数据
%matplotlib inline
import matplotlib.pyplot as plt
# 创建Figure实例
fig = plt.figure(figsize=(25,8),dpi=100)
# 添加子图
fig.add_subplot(2, 2, 1)
survived=pd.DataFrame({"survived":train["Survived"]})
plt.hist(survived)
fig.add_subplot(2, 2, 2)
pclass=pd.DataFrame({"pclass":train["Pclass"]})
plt.hist(pclass)
fig.add_subplot(2, 2, 3)
age=pd.DataFrame({"age":train["Age"]})
plt.hist(age)
fig.add_subplot(2, 2, 4)
train.loc[train["Embarked"] == "S", "Embarked"] = 1
train.loc[train["Embarked"] == "C", "Embarked"] = 2
train.loc[train["Embarked"] == "Q", "Embarked"] = 3
embarked=pd.DataFrame({"embarked":train["Embarked"]})
plt.hist(embarked)
#我通过多子图画法显示不同标签不同人的人数，
#生存状态，死亡人数远多于生存人数
#船舱人数，三号舱数最多
#泰坦尼克号人年龄基本上集中在20岁到40之间
#在s港登船的人数最多，不同登船地方的人，可能存活率不同
```

```python
Survived_train =pd.DataFrame(train.pop("Survived"))
Survived_train
```

```python
all_df = pd.concat((train, test), axis=0)
all_df.head()
```

```python
#对总集缺省值处理
all_df.isnull().sum().sort_values(ascending=False)
```

```python
#因为票价和船舱都是固定值，所以我认为应该用众数填充
train1=all_df["Embarked"]
train1.unique()
train1.value_counts()
```

```python
all_df['Embarked'] = all_df['Embarked'].fillna('S')
all_df.loc[all_df["Embarked"] == "S", "Embarked"] = 1
all_df.loc[all_df["Embarked"] == "C", "Embarked"] = 2
all_df.loc[all_df["Embarked"] == "Q", "Embarked"] = 3
all_df['Fare'] = all_df['Fare'].fillna('8.0500')
all_df.isnull().sum().sort_values(ascending=False)
```

```python
#对于船舱的缺省值，我认为不一定所以人都住得起，可能没有值的人就是住不起船舱的人
#所以我直接将船舱数据分为两种，0和1
all_df['Cabin'] = all_df['Cabin'].fillna('0')
all_df.head()
```

```python
all_df.loc[all_df["Cabin"] != "0", "Cabin"] = 1
all_df.head()
```

```python
all_df.isnull().sum().sort_values(ascending=False)
```

```python
#建立模型
#将数据集分回训练集和测试集
train_df = all_df.loc[train.index]
test_df = all_df.loc[test.index]
train_df.head()
```

```python
train_all=pd.merge(train_df,Survived_train,on="PassengerId") 
train_all.loc[train_all["Sex"] == "male", "Sex"] = 1
train_all.loc[train_all["Sex"] == "female", "Sex"] = 0
```

```python
df = dict([x for x in train_all.groupby('Survived')])[1]
df1 = dict([x for x in train_all.groupby('Survived')])[0]
Fare_df=df["Fare"].astype(float)
x1=Fare_df.drop_duplicates()
Fare_df1=df1["Fare"].astype(float)
y1=Fare_df1.drop_duplicates()
```

```python
x=df.groupby('Fare').count()
y=df1.groupby('Fare').count()
```

```python
x2=x['Pclass']
y2=y['Pclass']
plt.figure(figsize=(25,8))
plt.bar(x1,x2,width=0.3,color='plum',label='生存')
plt.bar(y1,y2,width=0.3,color='g',label='死亡')
plt.show()
#根据该表数据我发现，票价最低的人，属于绿色死亡的人占比很多，几乎全部死亡
#但是票价最高的存活确比死亡高，这说明资本还是存活的可能性高一些
#但是也绝不是肯定的，在一百以内也出现了存活显然高于死亡的部分，我认为可能和年龄或者性别有关
```

```python
Sex_df=df["Sex"]
s=Sex_df.drop_duplicates()
a=df.groupby('Sex').count()
b=df1.groupby('Sex').count()
a1=a['Pclass']
b1=b['Pclass']
plt.figure(figsize=(25,8),dpi=100)
plt.bar(s,a1,width=0.1,color='plum',label='生存')
plt.bar([i+0.1 for i in s],b1,width=0.1,color='g',label='死亡')
plt.xticks(range(0, 2),['female','male'])
plt.show()
#这是关于性别的存活率显示，可以很明显的看出来
#女性的存活率显然高于男性，女性存活率搞，男性死亡率高
```

```python
Age_df=df["Age"]
s1=Age_df.drop_duplicates()
Age_df1=df1["Age"]
s2=Age_df1.drop_duplicates()
z=df.groupby('Age').count()
t=df1.groupby('Age').count()
z1=z['Pclass']
t1=t['Pclass']
plt.figure(figsize=(25,8),dpi=100)
plt.bar(s1,z1,width=0.1,color='plum',label='生存')
plt.bar([i+0.1 for i in s2],t1,width=0.1,color='g',label='死亡')
plt.show()
#如图可以看出，十岁以下的孩子存活率最高，特别是不到一岁的孩子基本上存活
#在六十以上的老人基本死亡，但是出现一个异常值，八十岁存在存活，所以我使用箱型图再次观察
#在十五这一年龄存在全部死亡的情况，我认为是异常值，可能是因为年龄值的处理出现问题
```

```python
x_train = train_df.drop(['Ticket'], axis=1)
x_test = test_df.drop(['Ticket'], axis=1)
y_train = Survived_train
y_test = Survived_train
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction import DictVectorizer
dict = DictVectorizer(sparse=False)
X_train = dict.fit_transform(x_train.to_dict(orient='record'))
X_test = dict.transform(x_test.to_dict(orient='record'))
dict.get_feature_names()
X_train
```

```python
from sklearn.tree import DecisionTreeClassifier
dec = DecisionTreeClassifier(max_depth=5)
dec.fit(X_train, y_train)
```

```python
y_predict = dec.predict(X_test).astype(np.int64)
```

```python
test1  = pd.read_csv(r'C:/Users/LENOVO/Downloads/test (1).csv')
id = test1['PassengerId']
sub = {'PassengerId': id, 'Survived': y_predict}
submission = pd.DataFrame(sub)
submission
```
